```python
#1. Librerías.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
#2. Constantes.
df_final_path = "./pruebas_batch/df_final_IL_1610_1500samples.csv"
```

```python
#3. Lectura.
df_final = pd.read_csv(df_final_path)
```

```python
#4. Análisis estadísticos univariados básicos - Tabla.
df_final.describe()
```

```python
#5. Análisis estadísticos univariados básicos - Gráfico.
#a. Configuración general de gráficos.
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)

#b. Seleccion que columnas voy a graficar.
df_final = df_final.iloc[:,6:] 

#c. Función para analizar columnas automáticamente según tipo.
def analizar_columnas(df_final):
    for col in df_final.columns:
        print(f"\n{'='*40}\nColumna: {col}")
        print(df_final[col].describe(include='all'))

        if df_final[col].dtype == "object": 
            #i. Frecuencia de valores.
            counts = df_final[col].value_counts(dropna=False)
            print(counts)
            #ii. Gráfico de barras.
            #counts.plot(kind='bar')
            #plt.title(f"Frecuencia de '{col}'")
            #plt.ylabel("Conteo")
            #plt.show()
    
        elif df_final[col].dtype == "bool":
            #i. Frecuencia de valores.
            counts = df_final[col].value_counts(dropna=False)
            print(counts)
            #ii. Gráfico de barras.
            counts.plot(kind='bar')
            plt.title(f"Frecuencia de '{col}'")
            plt.ylabel("Conteo")
            plt.show()

        elif pd.api.types.is_numeric_dtype(df_final[col]):
            #iii. Estadísticas básicas.
            print(df_final[col].describe())
            #iv. Histograma.
            sns.histplot(df_final[col].dropna(), kde=True, bins=20)
            plt.title(f"Distribución de '{col}'")
            plt.xlabel(col)
            plt.show()

#c. Llamo a la función.
analizar_columnas(df_final)
```
